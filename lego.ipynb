````python
#year
def theme_by_year(year):
    #rename the columns in the themes column to be able to merge with the sets column
    data_rename = themes.rename(columns={"id": "theme_id", "name":"name_themes"})
    years = sets[sets['year'] == year]
    data_merge = pd.merge(data_rename, years, on='theme_id')
    #remove duplicates 
    unique_data = data_merge[['theme_id', 'name_themes']].drop_duplicates()
    #change back the theme_id to id
    unique_name = unique_data.rename(columns={'theme_id':'id'})
    data_reset = unique_name.reset_index(drop=True)
    return data_reset
````

**Q14:** 
- Create a function called `get_minifigs` that takes as input a set number (as a string) and a version number (as an integer) and outputs all minifigs from that set / version combination.  
- The DataFrame should be in `fig_num` order with the following columns/column names: `set_num`, `fig_num`, `fig_name`, `fig_num_parts`.  
- The index should go from 0 to n-1.

````python
minifigs
````

````python
def get_minifigs(set_num, version):
    #rename columns to correctly identify them
    rename_inventories = inventories.rename(columns={'id':'inventory_id'})
    rename_minifigs = minifigs.rename(columns={'name':'fig_name','num_parts':'fig_num_parts'})
    #merge data with respective columns
    mergeddata1 =pd.merge(rename_inventories, inventory_minifigs, on='inventory_id')
    mergeddata2 = pd.merge(mergeddata1, rename_minifigs, on='fig_num')
    #extract the set_num and version
    minifigs1 = mergeddata2[(mergeddata2['set_num']== set_num) & (mergeddata2['version']== version)]
    minifigs1 = minifigs1[['set_num','fig_num','fig_name', 'fig_num_parts']].sort_values(by='fig_num')
    minifigs1 = minifigs1.reset_index(drop=True)
    return minifigs1
````

**Q17:** 
- Create a function called `sets_from_minifig` that takes as input a minifig number (as a string) and returns a list of sets a minifig has appeared in.  
- Return a DataFrame (sorted by `set_num` and then by `version`) with the following columns: `set_num`, `version`,`name_sets`, `fig_num`, `fig_name`.  
- The index should go from 0 to n-1.

````python
def sets_from_minifig(fig_num):
    #rename appropiate columns
    themerenamed = themes.rename(columns={'id':'theme_id','name':'theme_name'})
    minifigsrenamed = minifigs.rename(columns ={'name':'fig_name','num_parts':'fig_num_parts'})
    inventoriesre = inventories.rename(columns={'id':'inventory_id'})
    setsrenamed = sets.rename(columns={'name':'name_sets','num_parts':'set_num_parts'})
    #merge all data together, with respect to columns in column
    data1 = pd.merge(themerenamed,setsrenamed, on='theme_id')
    data2 = pd.merge(inventoriesre, inventory_minifigs, on='inventory_id')
    data3 = pd.merge(data2,minifigsrenamed, on='fig_num')
    data4 = pd.merge(data3,data1, on='set_num')
    data4 = data4[data4['fig_num']== fig_num]
    data4 = data4[['set_num','version','name_sets','fig_num','fig_name']].sort_values(by=['set_num','version']).reset_index(drop=True)
    return data4
````

- Create a function called `set_parts` that takes as input the set number (as a string) and version number (as an integer) and outputs a DataFrame with each part in the set.  
- The DataFrame should have the following columns/column names: `set_num`, `version`, `name_sets`, `year`, `name_themes`, `part_num`, `name_part`, `name_color`,`quantity`,`is_spare`.  
- The Data should be sorted by `part_num` and then `name_color`.  
- The index should go from 0 to n-1.

````python
def set_parts(set_num,version):
    #rename appropiate columns
    renasets = sets.rename(columns={'name':'name_sets'})
    renathemes = themes.rename(columns={'name':'name_themes','id':'theme_id'})
    renaparts = parts.rename(columns={'name':'name_part'})
    renacolors = colors.rename(columns={'id':'color_id','name':'name_color'})
    renainventories = inventories.rename(columns={'id':'inventory_id'})
    #merge dataframes by columns
    data1 = pd.merge(renasets, renathemes, on='theme_id')
    data2 = pd.merge(data1, renainventories, on='set_num')
    data3 = pd.merge(data2,inventory_parts, on='inventory_id')
    data4 = pd.merge(data3, renaparts, on='part_num')
    data5 = pd.merge(data4, renacolors, on='color_id')
    data5 = data5[(data5['set_num']== set_num) & (data5['version']== version)]
    data6 = data5[['set_num','version','name_sets','year','name_themes','part_num','name_part','name_color','quantity','is_spare']]
    data6 = data6.sort_values(by=['part_num','name_color'])
    data6 = data6.reset_index(drop=True)
    return data6
````
